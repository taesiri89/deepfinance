## 전략 수정 : 전전전일 상승, 전전일 상승, *전일 하락*으로 수정 + 3일후 청산으로 정정

```python
buydate = datetime.datetime(2017,1,4)
retu = 1
buy = 0
tcnt = 0
position = 0
arretu = []
arrday = []
```

```python
for date in c.index:
    tcnt += 1
    
    if tcnt > 4 and position == 0 :   #매수
        bcode = []
        predate = c.index[tcnt-4]
        for code in c.columns:
            if r_c.ix[predate:date][code][-2]<0 and r_c.ix[predate:date][code][-3]>0 and r_c.ix[predate:date][code][-4]>0 :
                bcode.append(code)
                position =1 #포지션 보유
        
        buydate = date #매수일 저장
    
    if position == 1 and date>= buydate+datetime.timedelta(days=3):  #매도
        reball = []
        pfret = 0
        position = 0 #포지션 청산
        rebalretu = 0
        
        for buy in bcode:
            rebalretu = c.ix[date][buy]/c.ix[buydate][buy]-1  #수익률 계신
            reball.append(rebalretu)
            
        if len(bcode)>0:
            pfret = sum(reball)/len(bcode) #포트폴리오 동일가중 수익률 계산
            retu = retu*(1+pfret) #누적수익률 계산
            arretu.append(retu)
            arrday.append(date)
            print(date, 'cumulative_return', retu,'count',len(bcode),bcode)
```

```python
print('total_return',retu)
tret = pd.DataFrame(arretu, index=arrday)
plt.rcParams['lines.linewidth'] = 5
plt.rcParams['axes.grid'] = True
plt.figure(figsize=(10, 5))
plt.plot(tret)
plt.title('KRX300 momentum strategy')
plt.show()
```
