```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup
```

body > table.type2 > tbody > tr:nth-child(3) > td:nth-child(1) > span

```python
date = []
cl_price = []
st_price = []
hi_price = []
low_price = []
vol =[]
```

```python
for i in list(range(1,9)):
    req = requests.get("http://finance.naver.com/item/sise_day.nhn?code=263750&page="+str(i))
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')

    my_prices = soup.select('tr > td > span')

    price_info = []
    for price in my_prices:
        if len(price.text) > 1:
            price_info.append(price.text)

    for j in list(range(0,len(price_info)-5,6)):
        date.append(price_info[j])
        cl_price.append(price_info[j+1])
        st_price.append(price_info[j+2])
        hi_price.append(price_info[j+3])
        low_price.append(price_info[j+4])
        vol.append(price_info[j+5])
```

```python
pearl = pd.DataFrame()
pearl['date'] = pd.Series(date[::-1])
pearl['cl_price'] = pd.Series(cl_price[::-1])
pearl['st_price']=pd.Series(st_price[::-1])
pearl['hi_price']=pd.Series(hi_price[::-1])
pearl['low_price']=pd.Series(low_price[::-1])
pearl['vol'] = pd.Series(vol[::-1])
```
